```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
text = ['the glass of milk',
       'the glass of water',
       'I prefer coffee than tea',
       'I love travelling from one place to another',
       'I love machone learning specially data science']

text
```

# Using One_Hot

```python
voc_size=100
one_hot_repre = [one_hot(words, voc_size) for words in text]
one_hot_repre
```

# Using Embedding

```python
padding_sequence = pad_sequences(one_hot_repre, padding='post', maxlen=10)
print(padding_sequence)
```

```python
model = Sequential()
model.add(Embedding(100, 8, input_length=10))

#100 for number of distinct word or vocabulary size
#8 for size of the embedding vector
#input_length=10 referes the size of each input sequence.
```

```python
model.summary()
```

```python
model.compile(loss='mse',
             optimizers='adam',
             metrics=['accuracy'])
```

```python
model.predict(padding_sequence)
```
